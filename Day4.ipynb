```python
test_scores = [72, 95, 58, 84, 41, 89, 92]
i=max(test_scores)
j=min(test_scores)
k=i-j
print(f"最大値は{i},最小値は{j},範囲は{k}")
if k > 50:
    print("High Variance")
else:
    print("Consistent")
```
