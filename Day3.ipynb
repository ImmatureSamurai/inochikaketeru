```python
scores = [85, 42, 78, 92, 55, 60, 38, 74, 88]
passed_scores=[]
sum=0
ave=0
j=0
for i in scores :
  if i>=60:
    passed_scores.append(i)
    sum=sum+i
    j=j+1
ave=sum/j
print(ave)
if j>=5:
  print("Class performance is good")
else:
  print("Needs improvement")
```
