```python
import random

print("""
VOCAB APP
Enter: next
finish: stop mode
exit: quit
""")

words = []

try:
    with open("words.txt", "r") as f:
        for line in f:
            parts = line.strip().split(",")
            if len(parts) >= 2:
                words.append({"word": parts[0], "meaning": parts[1]})
except FileNotFoundError:
    pass


while True:

    mode = input("\nadd / quiz / delete / exit: ").strip().lower()

    if mode == "add":
        while True:
            word = input("Word: ")
            if word.strip().lower() == "finish":
                break

            meaning = input("Meaning: ")
            if meaning.strip().lower() == "finish":
                break

            if any(d["word"] == word for d in words):
                print("exists")
                continue

            words.append({"word": word, "meaning": meaning})

            if input().strip().lower() == "finish":
                break

        with open("words.txt", "w") as f:
            for i in words:
                f.write(i["word"] + "," + i["meaning"] + "\n")

    elif mode == "delete":
        while True:
            target = input("Delete: ")

            if target.strip().lower() == "finish":
                break

            before = len(words)
            words = [d for d in words if d["word"] != target]

            if len(words) == before:
                print("not found")
            else:
                print("deleted")

            if input().strip().lower() == "finish":
                break

        with open("words.txt", "w") as f:
            for i in words:
                f.write(i["word"] + "," + i["meaning"] + "\n")

    elif mode == "quiz":
        correct = 0
        total = 0
        wrong_words = []

        while len(words) > 0:

            i = random.choice(words)

            print("\n", i["word"])
            ans = input()

            if ans.strip().lower() == "finish":
                break

            if ans == i["meaning"]:
                print("ok")
                correct += 1
            else:
                print("ng", i["meaning"])
                wrong_words.append(i)

            total += 1

        print(correct, "/", total)

        if len(wrong_words) > 0:
            if input().strip().lower() == "y":

                cr = 0
                tt = 0

                while len(wrong_words) > 0:

                    i = random.choice(wrong_words)

                    print("\n", i["word"])
                    ans = input()

                    if ans.strip().lower() == "finish":
                        break

                    if ans == i["meaning"]:
                        print("ok")
                        wrong_words.remove(i)
                        cr += 1
                    else:
                        print("ng", i["meaning"])

                    tt += 1

                print(cr, "/", tt)

    elif mode == "exit":
        break
```
